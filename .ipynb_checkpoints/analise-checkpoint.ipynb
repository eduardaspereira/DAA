```python
dt = pd.get_dummies(dt, columns=['LUMINOSITY'], prefix='Luminosity')
```

```python
dt['record_date_dt'] = pd.to_datetime(dt['record_date'], format='mixed', dayfirst=True)
dt['Hora'] = dt['record_date_dt'].dt.hour
dt['DiaSemana'] = dt['record_date_dt'].dt.dayofweek
dt['Mes'] = dt['record_date_dt'].dt.month
dt = dt.drop(columns=['record_date', 'record_date_dt'])
```
