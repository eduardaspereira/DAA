```python
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10,6)
# Gráfico de Contagem (Countplot) para a variável alvo
plt.title('Distribuição das Classes de Diferença de Velocidade (Alvo)')
sns.countplot(data=df_train, x='AVERAGE_SPEED_DIFF', order=df_train['AVERAGE_SPEED_DIFF'].value_counts().index)
plt.xlabel('Categoria de Diferença de Velocidade')
plt.ylabel('Contagem')
plt.show()
```

```python
df_train['record_date'] = pd.to_datetime(df_train['record_date'])
df_test['record_date'] = pd.to_datetime(df_test['record_date'])

# Extrair features temporais para ambos os datasets
for df in [df_train, df_test]:
    df['hora_do_dia'] = df['record_date'].dt.hour
    df['dia_semana'] = df['record_date'].dt.dayofweek # (Segunda=0, Domingo=6)
    df['mes'] = df['record_date'].dt.month
    df['fim_de_semana'] = df['dia_semana'].isin([5, 6]).astype(int) # 1 se Fim de Semana, 0 caso contrário
```
